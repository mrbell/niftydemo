```python
def setup(whatever_space):
    
    # e.g. the fourier domain for our given space
    k_space = whatever_space.get_codomain()

    # We define an isotropic, homogeneous random field as our signal
    # using the following power spectrum
    power = (lambda k: 42 / (k + 1) ** 3)
    S = nifty.power_operator(k_space, spec=power)
    s = S.get_random_field(domain=whatever_space)

    # define response, just a unit response in this case
    R = nifty.response_operator(whatever_space, sigma=0.0, mask=1.0, assign=None) 
    # this is "data space"
    d_space = R.target                                               

    # Define some noise variance; e.g., signal-to-noise ratio of 1
    N = nifty.diagonal_operator(d_space, diag=s.var(), bare=True)
    # generate noise
    n = N.get_random_field(domain=d_space)                           
    
    # compute data
    d = R(s) + n                                                    
    
    return d, s, S, N, R
    
def plot_stuff(s, d, m):
    s.plot(title="signal")
    d_ = nifty.field(s.domain, val=d.val, target=s.domain.get_codomain())
    d_.plot(title="data", vmin=s.min(), vmax=s.max())                
    m.plot(other=s, title="reconstructed map", vmin=s.min(), vmax=s.max())  
```

Now define a Wiener Filter function. Recall the Wiener Filter formula:

$m = Dj$

$m = (S^{-1} + R^\dagger N^{-1} R)^{-1} R^\dagger N^{-1}d$

```python
def wiener_filter(d, S, N, R):
    j = R.adjoint_times(N.inverse_times(d))
    D = nifty.propagator_operator(S=S, N=N, R=R) # Has the form of D, which appears often

    m = D(j, W=S, tol=1E-3, note=True)  # uses a conjugate gradient to multiply D with j
    
    return m
```

Let's apply the filter in 1D...

```python
space1 = nifty.rg_space(512)

(d1, s1, S1, N1, R1) = setup(space1) 
m1 = wiener_filter(d1, S1, N1, R1)
```

```python
plot_stuff(s1, d1, m1)
```

now in a 2D cartesian space...

```python
space2 = nifty.rg_space([128, 128])

(d2, s2, S2, N2, R2) = setup(space2) 
m2 = wiener_filter(d2, S2, N2, R2)
```

```python
plot_stuff(s2, d2, m2)
```

and lastly in a HealPix gridding of a spherical geometry

```python
space3 = nifty.hp_space(32)

(d3, s3, S3, N3, R3) = setup(space3) 
m3 = wiener_filter(d3, S3, N3, R3)
```

```python
plot_stuff(s3, d3, m3)
```
